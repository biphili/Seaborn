```python
import pandas as pd 
```

```python
import seaborn as sns 
```

```python
iris=pd.read_csv('Iris_1.csv')
```

```python
iris.head()
```

```python
iris['Species'].value_counts()
```

## Joint Plot

```python
%matplotlib inline
```

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris,size=5)
```

```python
sns.jointplot(x='PetalLengthCm',y='PetalWidthCm',data=iris,size=5)
```

## FacetGrid Plot

```python
import matplotlib.pyplot as plt 
```

```python
sns.FacetGrid(iris,hue='Species',size=5)\
.map(plt.scatter,'SepalLengthCm','SepalWidthCm')\
.add_legend()
```

## Box Plot

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
sns.boxplot(x='Species',y='PetalWidthCm',data=iris)
```

```python
sns.boxplot(x='Species',y='SepalWidthCm',data=iris)
```

```python
sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
```

```python
ax=sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray')
```

## Strip plot

```python
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray')
```

## Violin PLot

```python
sns.violinplot(x='Species',y='SepalLengthCm',data=iris,size=6)
```

```python
sns.pairplot(data=iris,kind='scatter')
```

```python
sns.pairplot(iris,hue='Species')
```
