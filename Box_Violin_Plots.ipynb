This is a tutorial on how to plot box plots and Violin plots in Python using Seaborn module

```python
# Normal Modules 
import numpy as np
from numpy.random import randn
import pandas as pd

# Import stats Library from numpy
from scipy import stats

# Modules for plotting 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 1.Box Plots 

```python
data1=randn(100)
data2=randn(100)
```

```python
sns.boxplot(data1)

# For plotting two plots together 
# sns.boxplot(data1,data2) 

# To include outliers in the wisker 
# sns.boxplot(data1,data2,whis=np.inf)

# Making the box plot horizontal 
# sns.boxplot(data1,vert=False)
```

# 2.Violin Plots

```python
# Normal Distribution 
data1=stats.norm(0,5).rvs(100)

# Two Gamma istributions and Concatenate together 
data2=np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])

# Box plots for data1 and data2
sns.boxplot(data1,whis=np.inf)
```

```python
sns.violinplot(data1,color='red')
```

```python
# Specifying Band width for Violin Plot
sns.violinplot(data1,bw=0.01)
```

```python
# Representing individual datapoints in a Violin Plot 
sns.violinplot(data1,inner='stick')
```
